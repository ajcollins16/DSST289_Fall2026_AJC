## Google Colab Setup

You might notice some weird "quarks" with google colab that I feel is best to change to help with the experience. When are going to change some of the default settings to help with things readability and/or visability. Specifically:

* under Editor change "Indentation width in spaces" to 4.
* under Editor, check "show line numbers".
* Under AI Assistance, click the "Hide generative AI features" and unclick everything else (very important!)

some other ones if your more of a classic "programmer" you can also do (I don't since I like to have some things the way they are):

* Under Editor, unclick "Automatically close brackets..."
* Under Editor, under "Code Diagnostics" select "None".

There are also two very important keyboard shortcuts to know:

- **Run selected cell in place**: Cmd + Enter (macOS), Ctrl + Enter (Windows)
- **Run cell and move to next cell**: Shift + Enter (macOS and Windows)

A few other things are helpful to know about Colab. Under File, you can click on "Save a Copy in Drive". This will save your work in a special Colab folder within your Google Drive. Otherwise, you will loss all of your work when closing the window.

If you reopen the Notebook from your Google Drive, it will use your default language settings, which may be difficult to use if it is not in English. You can change this by selecting Help (always the last item in the menu) and "Show in English" (always the last item in the pull-down menu).
