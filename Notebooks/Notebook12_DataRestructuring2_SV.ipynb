Today we will be using the criterion films dataset since it already has multiple sub-dataframes within it keyed on the movie itself. Many times data is collected and stored in many different types of. For example, in my field of psychology our unit of observation is people where participants are the rows and the columns are observations we have made. But in my specific field of experimental psychology trials are the unit of observation and a single participant is the entire dataset. Therefore, the data we collect can be organized in a way that is not good for visualization, analysis, or both. So we need ways to pivot or restructure the data to make it better for visualization, analysis, or both. The two most common are the pivot and unpivot. Now bare with me since I come from the R programming language where pivot is called pivot wider and unpivot is called pivot longer. But the idea is the same:

- Unpivot (Pivot longer) -> taking a wide tabular of data and making it wider.
- Pivot (Pivot wider) -> taking a long tabular of data and making it wider.

As an example for unpivot, a sales table has one row per store, with separate columns for monthly sales (`januarySales`, `FebruarySales`). We want to unpivot this from wide to long form by turning those month columns into identifiable (categorical) variables in a columns. So not the longer tabular is one with sales and month as the only columns. Pivot is then the opposite of this.

Finally, we also have the unnesting type of restructuring where data that is put together into a single row is expanded out or exploded.
