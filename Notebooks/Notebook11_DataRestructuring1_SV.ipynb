## Data Restructuring — Joining

Most questions that need answering with any form of data analysis is going to require the joining or pooling of datasets together to get a more accurate or full picture. This way a data scientists can do their best to answer the question that they are being asked. Therefore we need the ability to join data based on a *foreign key* relationship:

- Primary Key -> A column (or columns) that identify each row of a dataset
    - Candidate Key -> a column (or columns) that could be used as a primary key.
- Foreign Key -> A column (or columns) in a different dataset that also has the primary key

To then join these similar datasets we use the `.join()` function from the `polars` package which joins based on this primary and foreign key relationship. There are then seven kinds of joins:

- Inner join -> Keeps only rows with matching keys in both tables
- Left/Right join -> Keeps all rows from the left/right table, adding matching information from the right/left table when available.
- Full (Outer) join -> Keeps all rows from both tables, filling unmatched values with `null`.
- Semi join -> Keeps rows from the left table that have a match in the right table, but does not add columns from the right table.
- Anti join -> Keeps rows from the left/right table that do not have a match in the right/left table
- Equi-join -> A join based on equality, meaning rows are matched when key values are the same across tables.
- Cross join -> join everything disregarding the key.
