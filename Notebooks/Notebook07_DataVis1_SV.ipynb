## Grammer of Graphics and Scatter Plots

In this section, and for several notebooks, we review how to use the `plotnine` package to create visualizations in Python using the grammar of graphics. The grammar of graphics treats plots as a combination of layers, where each layer connects elements of a dataset to visual features of a plot. To build a visualization, we first specify the dataset, then define the aesthetic mappings that link columns in the data to parts of the graph, such as the x-axis, y-axis, color, shape, or size. Finally, we add a geometry, which determines what kind of visual object will represent each observation. We introduce these topics and ideas with a simple scatter plot between two continuous variables.
